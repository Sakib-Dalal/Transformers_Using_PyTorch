### Simplified encoder layer example

<img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Fwww.researchgate.net%2Fpublication%2F334288604%2Ffigure%2Ffig1%2FAS%3A778232232148992%401562556431066%2FThe-Transformer-encoder-structure.ppm&f=1&nofb=1&ipt=41cff290b603e54651f280652c0c36716a4db949384a706382450d2c4de21547">

```python
class EncoderLayer(nn.Module):
    def __init__(self, d_model, nhead, dim_feedforward, dropout=0.1, mask=None):
        super().__init__()

        # Self Attention Layer
        self.self_attn = nn.MultiheadAttention(embed_dim=d_model, num_heads=nhead, dropout=dropout, device=DEVICE, batch_first=True)

        # FFN Layer
        self.feed_forward = nn.Sequential(
            nn.Linear(d_model, 2 * dim_feedforward),
            nn.GLU(dim=-1), # shape: dim_feedforward
            nn.Linear(in_features=dim_feedforward, out_features=d_model),
            nn.Dropout(dropout),
        )

        # Normalization Layers
        self.norm1 = nn.LayerNorm(normalized_shape=d_model)
        self.norm2 = nn.LayerNorm(normalized_shape=d_model)

        # Dropout layers
        self.dropout1 = nn.Dropout(dropout)
        self.dropout2 = nn.Dropout(dropout)

    def forward(self, x, mask=None):
        # Self Attention Layer
        attn_output, attn_output_weights = self.self_attn(query=x, key=x, value=x, attn_mask=mask)
        
        x = x + self.dropout1(attn_output) # -------> Residual connection
        x = self.norm1(x) # ---------> Normalisation

        # FFN Layer
        ff_output = self.feed_forward(x)
        
        x = x + self.dropout2(ff_output) # -------> Residual connection
        x = self.norm2(x) # ---------> Normalisation

        return x, attn_output_weights
```

```python
# Parameters
d_model = 512
nhead = 1
dim_feedforward = 1024
mask = None
dropout=0.1
```

```python
encoder = EncoderLayer(d_model=d_model, nhead=nhead, dim_feedforward=dim_feedforward, dropout=dropout, mask=mask).to(DEVICE)
```

### Scaled Dot-Product Attention
<img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Ftse1.mm.bing.net%2Fth%2Fid%2FOIP.ViTTK6Ufe6aKZxOIRdx9TwHaCB%3Fpid%3DApi&f=1&ipt=ec547870ed2e4bab69bafc15f4e016afe8d17ab975edc24005d162c363dc8bd4&ipo=images">

<img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Fwww.researchgate.net%2Fpublication%2F349847862%2Ffigure%2Ffig2%2FAS%3A1002388374691840%401615999420067%2FScaled-dot-product-attention-This-picture-introduces-the-structure-of-the-Scaled.png&f=1&nofb=1&ipt=e98578a6172a9802d774887811a013810ac1f00547d4d15ee8a67f9a905b76f9" style="height:500px">

```python
import torch.nn.functional as F
```

```python
class ScaledDotProductAttention(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, Q, K, V, mask=None):
        """
        Q: (batch_size, seq_len, d_model)
        K: (batch_size, seq_len, d_model)
        V: (batch_size, seq_len, d_model)
        """

        d_k = K.size(-1)

        # 1. QK^T
        matmul = torch.matmul(Q, K.transpose(-2, -1))
        scale = matmul / math.sqrt(d_k)

        # 2. Mask (not in encoder)
        if mask is not None:
            scale = scale.masked_fill(mask == 0, float('-inf'))

        # 3. Softmax
        attention_weights = F.softmax(input=scale, dim=-1)

        # 4. Weighted sum
        output = torch.matmul(attention_weights, V)

        return output, attention_weights
```

```python
X = torch.randn(1, 10, 512)
X.shape
```

```python
sdpa = ScaledDotProductAttention().to(DEVICE)
```

```python
y = sdpa(X, X, X, mask=None)
```

## Multi-Head Attention

$d_k = d_v = d_{\text{model}}/h=64$

where $h$=8 and $d_k$ refers to the dimension of the keys and $d_v$ refers to the dimension of the values and $d_{\text{model}}$ to the model's dimension.

<img src="https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Fwww.jeremyjordan.me%2Fcontent%2Fimages%2F2023%2F05%2Fmulti-head-attention.png&f=1&nofb=1&ipt=9f250554f4be67e52e94a698a13d2ffe16f396b44dc8e485cc5cf0df0b3807d8" style="height:400px">

```python
class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads):
        super(MultiHeadAttention, self).__init__()

        assert d_model % num_heads == 0 # check d_model is divisible by num_heads and their reminder is 0

        self.d_model = d_model
        self.num_heads = num_heads

        self.d_k = d_model // num_heads

        # Define Q, K, V, Output Linear Layers
        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)

        self.W_o = nn.Linear(d_model, d_model)

    def forward(self, query, key, value, mask=None):
        batch_size = query.size(0)

        # for Q -> (batch_size, num_heads, sequence_length, depth)
        query = self.W_q(query)
        Q = query.view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)

        # for K -> (batch_size, num_heads, sequence_length, depth)
        key = self.W_k(key)
        K = key.view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)

        # for V -> (batch_size, num_heads, sequence_length, depth)
        value = self.W_k(value)
        V = value.view(batch_size, -1, self.num_heads, self.d_k).transpose(1, 2)

        '''
            Scaled Dot Poduct Attention START
        '''
        # all scores from each heads 
        scores = torch.matmul(Q, K.transpose(-2, -1))
        scores_norm = scores / math.sqrt(self.d_k)

        if mask is not None:
            scores_norm = scores_norm.masked_fill(mask==0, -1e9)

        # applying softmax
        attention = F.softmax(scores_norm, dim=-1)

        output = torch.matmul(attention, V)
        # Output -> (batch_size, sequence_length, embedding_dimension)
        output = output.transpose(1, 2).contiguous().view(batch_size, -1, self.d_model)

        output = self.W_o(output)

        return output
```

```python
# Assuming we have some data
batch_size = 32
sequence_length = 100
d_model = 512
num_heads = 8
mask = None

# Instantiate the model
multi_head_attn = MultiHeadAttention(d_model=d_model, num_heads=num_heads)
multi_head_attn
```

```python
# Create some random data
input_data = torch.rand(batch_size, sequence_length, d_model)
input_data.shape
```

```python
# We use the same input for Q, K, V for self attention
output = multi_head_attn(input_data, input_data, input_data, mask=mask)
output.shape
```

### Positional Encoding (Transformer)

For even indices:

$$PE_{(\text{pos}, 2i)} = \sin\left(\frac{\text{pos}}{10000^{\frac{2i}{d_{\text{model}}}}}\right)$$

For odd indices:

$$PE_{(\text{pos}, 2i+1)} = \cos\left(\frac{\text{pos}}{10000^{\frac{2i}{d_{\text{model}}}}}\right)$$


```python
num_hiddens = 768

class PositionalEncoding(nn.Module):
    '''Positional encoding class.'''
    def __init__(self, num_hiddens, max_len=10000):
        super().__init__()

        self.dropout = nn.Dropout(0.1)

        # Create a positional embedding matrix
        position = torch.arange(max_len, dtype=torch.float32).reshape(-1, 1)

        div_term = torch.exp(
            torch.arange(0, num_hiddens, 2, dtype=torch.float32) * -(math.log(10000.0) / num_hiddens)
        )

        pe = torch.zeros((1, max_len, num_hiddens))
        pe[0, :, 0::2] = torch.sin(position * div_term)
        pe[0, :, 1::2] = torch.cos(position * div_term)

        self.register_buffer('pe', pe)

    def forward(self, X):
        X = X + self.pe[:, :X.shape[1], :].to(X.device)
        return self.dropout(X)
```

```python
pos_enc = PositionalEncoding(num_hiddens=num_hiddens, max_len=10000)
X_pos = pos_enc(X)
```
